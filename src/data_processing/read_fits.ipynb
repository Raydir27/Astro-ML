```python
from astropy.io import fits
from pathlib import Path

# Folder containing FITS files
fits_dir = Path("../../Data/XMM-Gaia-Cross-Matched Data Separate")

# Find all .fits files
fits_files = sorted(fits_dir.glob("*.fits"))

print(f"Found {len(fits_files)} FITS files\n")

MAX_COLS = 5

for fits_path in fits_files:
    print("=" * 100)
    print(f"FILE: {fits_path.name}")
    print("=" * 100)

    try:
        with fits.open(fits_path) as hdul:
            # Show HDU structure
            hdul.info()

            # Safely check if HDU 1 exists and is a table
            if len(hdul) > 1 and hasattr(hdul[1], "columns"):
                data = hdul[1].data
                columns = hdul[1].columns

                print("\nColumn name | Format | First-row value")
                print("-" * 80)

                for col in columns:
                    name = col.name
                    fmt = col.format
                    value = data[name][0] if len(data) > 0 else None
                    print(f"{name:30s} | {fmt:8s} | {value}")
            else:
                print("\nNo table found in HDU[1].")

    except Exception as e:
        print(f"Error reading {fits_path.name}: {e}")

    print("\n")  # spacing between files
```
