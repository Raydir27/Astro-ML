```python
from astropy.table import Table
import pandas as pd

def extract_and_filter():
    # 1. Load XMM Data
    xmm_path = r"C:\Raj_Stuff\Coding\VScode\Workspaces\Astro-ML\Data\XMM-Gaia-Cross-Matched Data Separate\Unique_SRCID_cluster_0_0_OBSID_all_SRCID.fits"
    xmm_table = Table.read(xmm_path)
    
    # Select specific XMM features
    # Note: Using 'No/Nx' exactly as written; ensure the column name matches your FITS header
    xmm_cols = [
        # 'SC_RA', 'SC_DEC', 
        'SC_POSERR', 'SC_DET_ML', 
        'N_DETECTIONS', 'CONFUSED', 'HIGH_BACKGROUND', 'No/Nx', 'Clusters'
    ]
    xmm_filtered = xmm_table[xmm_cols].to_pandas()

    # 2. Load Gaia Data
    gaia_path = r"C:\Raj_Stuff\Coding\VScode\Workspaces\Astro-ML\Data\XMM-Gaia-Cross-Matched Data Separate\gaiadr3_cluster_0_0.fits"
    gaia_table = Table.read(gaia_path)
    
    # Select specific Gaia features
    gaia_cols = [
        # 'gaia_RA_ICRS', 'gaia_DE_ICRS',
        'gaia_e_RA_ICRS', 'gaia_e_DE_ICRS', 
        'gaia_pmRA', 'gaia_e_pmRA', 'gaia_pmDE', 'gaia_e_pmDE', 
        'gaia_Plx', 'gaia_e_Plx'
    ]
    gaia_filtered = gaia_table[gaia_cols].to_pandas()

    return xmm_filtered, gaia_filtered

# Run extraction
try:
    xmm_df, gaia_df = extract_and_filter()
    print("XMM Features Preview:")
    print(xmm_df.head())
    print("\nGaia Features Preview:")
    print(gaia_df.head())
except KeyError as e:
    print(f"Error: One of the columns was not found in the FITS file: {e}")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
import pandas as pd

# 1. Clean XMM Data
# Decode byte-strings like b'Clus_0_0' to regular strings 'Clus_0_0'
if xmm_df['Clusters'].dtype == object:
    xmm_df['Clusters'] = xmm_df['Clusters'].str.decode('utf-8')

# Convert Booleans to Integers (0 and 1) for ML compatibility
xmm_df['CONFUSED'] = xmm_df['CONFUSED'].astype(int)
xmm_df['HIGH_BACKGROUND'] = xmm_df['HIGH_BACKGROUND'].astype(int)

# Rename the column with the slash to prevent syntax issues later
xmm_df.rename(columns={'No/Nx': 'No_Nx'}, inplace=True)

# 2. Clean Gaia Data
# Gaia often has NaNs in proper motion (pm) and parallax (Plx) for faint stars.
# Depending on your ML goal, you can fill with 0, mean, or drop them.
# For now, let's keep them but be aware they are there.

# 3. Save to Parquet (Preserves types and handles NaNs cleanly)
xmm_df.to_parquet('xmm_filtered_cleaned.parquet', engine='fastparquet', index=False)
gaia_df.to_parquet('gaia_filtered_cleaned.parquet', engine='fastparquet', index=False)

print("Data cleaned and saved to Parquet.")
```
